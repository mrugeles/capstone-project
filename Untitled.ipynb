```python
import numpy as np
import pandas as pd
from time import time
import matplotlib.pyplot as plt
from IPython.display import display 
from sklearn.metrics import fbeta_score, make_scorer

import model_utils as utils

# Pretty display for notebooks
%matplotlib inline
pd.set_option('display.max_colwidth', -1)

dataset, features, labels = utils.getDataSet("datasets/dataset.csv")
```

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pandas as pd
print(dataset.shape)
kf = KFold(n_splits=5, shuffle=True)
kf.get_n_splits(features)

clf = RandomForestClassifier(random_state = 300)
dfResults = pd.DataFrame(columns=['learner', 'train_time', 'pred_time', 'f_test', 'f_train'])

for train_index, test_index in kf.split(features):
    X_train, X_test = features.iloc[train_index], features.iloc[test_index]
    y_train, y_test = labels.iloc[train_index], labels.iloc[test_index]
    clf, dfResults = utils.train_predict(clf, X_train, y_train, X_test, y_test, dfResults)
    

```

```python
display(dfResults)
```

```python
from sklearn.utils import shuffle

features_shuffle, labels_shuffle = shuffle(features, labels, random_state=0)
```

```python
from sklearn.model_selection import cross_val_score
ftwo_scorer = make_scorer(fbeta_score, beta=2)

scores = cross_val_score(clf, features_shuffle, labels_shuffle, scoring = ftwo_scorer, cv=5)

print(scores)
print("Score: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
